```python
# Device agnostic
device = "mps" if torch.backends.mps.is_available() else "cpu"
```

```python
# Calculate accuracy
def accuracy(y_true, y_pred):
    correct = torch.eq(y_true, y_pred.round()).sum().item()
    acc = correct / len(y_true) * 100
    return acc
```

```python
# Build a model
class MulticlassModel(nn.Module):
    def __init__(self, num_features, output_features, hidden_units=256):
        super().__init__()
        # Define layers
        self.linear_layer_stack = nn.Sequential(
            nn.Linear(in_features=num_features, out_features=hidden_units),
            nn.ReLU(),
            nn.Linear(in_features=hidden_units, out_features=hidden_units),
            nn.ReLU(),
            nn.Linear(in_features=hidden_units, out_features=output_features)
            )
    # Define forward method
    def forward(self, X):
        return self.linear_layer_stack(X)

# Instantiate model class & send to target device
model_4 = MulticlassModel(num_features=55,
                          output_features=3
                          ).to(device)
model_4
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau

# Training Loop & Testing loop
# Set random seed
torch.manual_seed(42)

# Put data on device
X_train, X_test = X_train.to(device), X_test.to(device)
y_train, y_test = y_train.to(device), y_test.to(device)

# Loss and optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model_4.parameters(), lr=0.01)
scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=10, factor=0.1, verbose=True)

epochs = 1000

for epoch in range(epochs):
    model_4.train()
    
    # Forward pass
    y_logits = model_4(X_train)
    y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1) # logits -> probs -> labels

    # Calculate loss/accuracy
    loss = loss_fn(y_logits, y_train)
    train_acc = accuracy(y_true=y_train,
                         y_pred=y_pred)
    
    optimizer.zero_grad() # reset gradients to zero
    loss.backward() # backpropagation
    optimizer.step() # gradient descent

    model_4.eval()
    with torch.no_grad():  # Use torch.no_grad() instead of torch.inference_mode()
        
        test_logits = model_4(X_test)
        test_pred = torch.softmax(test_logits, dim=1).argmax(dim=1)

        test_loss = loss_fn(test_logits, y_test)
        # Update the learning rate based on the validation loss
        scheduler.step(test_loss)

        test_acc = accuracy(y_true=y_test,
                            y_pred=test_pred)
        
        if (epoch+1) % 100 == 0:
            print(f"Epoch: {epoch+1}, Loss: {loss:.5f}, Train Acc: {train_acc:.2f}%, Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")
```

```python
# Device agnostic
device = "mps" if torch.backends.mps.is_available() else "cpu"
```

```python
import torch
import torch.nn as nn

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=512):
        super().__init__()

        # Create a matrix of zeros with shape (max_len, d_model) to store positional encodings.
        self.encoding = torch.zeros(max_len, d_model)

        # Generate a sequence of numbers from 0 to (max_len-1) and reshape it to (max_len, 1) to represent positions.
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)

        # Calculate div_term using a logarithmic term to scale the positional encodings appropriately.
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-torch.log(torch.tensor(10000.0)) / d_model))

        # Apply sine function to even-indexed columns and cosine function to odd-indexed columns of the positional encoding matrix.
        self.encoding[:, 0::2] = torch.sin(position * div_term)
        self.encoding[:, 1::2] = torch.cos(position * div_term)

        # Add a batch dimension and transpose the positional encodings to have shape (max_len, 1, d_model).
        self.encoding = self.encoding.unsqueeze(0).transpose(0, 1)

    def forward(self, x):
        return x + self.encoding[:x.size(0), :]

class MultiHeadAttentionTransformer(nn.Module):
    def __init__(self, num_features, output_features, hidden_units=256, num_heads=4, dropout=0.1):
        super().__init__()

        # Linear layer for embedding the input features into the hidden_units dimension.
        self.embedding = nn.Linear(num_features, hidden_units)

        # PositionalEncoding module to add positional encodings to the input embeddings.
        self.positional_encoding = PositionalEncoding(hidden_units)

        # MultiheadAttention layer for performing multi-head self-attention on the input sequence.
        self.multihead_attention = nn.MultiheadAttention(hidden_units, num_heads, dropout=dropout)

        # Feed-forward neural network with two linear layers and ReLU activation.
        self.feed_forward = nn.Sequential(
            nn.Linear(hidden_units, hidden_units),
            nn.ReLU(),
            nn.Linear(hidden_units, hidden_units)
        )

        # Layer normalization after the first self-attention block.
        self.layer_norm1 = nn.LayerNorm(hidden_units)

        # Layer normalization after the feed-forward neural network block.
        self.layer_norm2 = nn.LayerNorm(hidden_units)

        # Dropout layer for the output of the first self-attention block.
        self.dropout1 = nn.Dropout(dropout)

        # Dropout layer for the output of the feed-forward neural network block.
        self.dropout2 = nn.Dropout(dropout)

        # Linear layer for the final output, mapping the hidden_units dimension to the number of output features.
        self.output_layer = nn.Linear(hidden_units, output_features)


    def forward(self, X):
        # Apply the embedding layer and add positional encodings
        X = self.embedding(X) * torch.sqrt(torch.tensor(X.size(-1), dtype=torch.float))
        X = self.positional_encoding(X)

        # Transpose for multi-head attention
        X = X.transpose(0, 1)

        # Multi-head attention
        attention_output, _ = self.multihead_attention(X, X, X)
        X = X + self.dropout1(attention_output)
        X = self.layer_norm1(X)

        # Feed-forward neural network
        feed_forward_output = self.feed_forward(X)
        X = X + self.dropout2(feed_forward_output)
        X = self.layer_norm2(X)

        # Transpose back and apply output layer
        X = X.transpose(0, 1)
        output = self.output_layer(X)

        return output

# Instantiate model class & send to target device
model_transformer = MultiHeadAttentionTransformer(num_features=55,
                                                  output_features=3
                                                  ).to(device)
model_transformer
```

```python
# Training Loop & Testing loop
# Set random seed
torch.manual_seed(42)

# Put data on device
X_train, X_test = X_train.to(device), X_test.to(device)
y_train, y_test = y_train.to(device), y_test.to(device)

# Loss and optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model_transformer.parameters(), lr=0.01)
scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=10, factor=0.1, verbose=True)

epochs = 1000

for epoch in range(epochs):
    model_transformer.train()
    
    # Forward pass
    y_logits = model_transformer(X_train)
    y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1) # logits -> probs -> labels

    # Calculate loss/accuracy
    loss = loss_fn(y_logits, y_train)
    train_acc = accuracy(y_true=y_train,
                         y_pred=y_pred)
    
    optimizer.zero_grad() # reset gradients to zero
    loss.backward() # backpropagation
    optimizer.step() # gradient descent

    model_transformer.eval()
    with torch.no_grad():  # Use torch.no_grad() instead of torch.inference_mode()
        
        test_logits = model_transformer(X_test)
        test_pred = torch.softmax(test_logits, dim=1).argmax(dim=1)

        test_loss = loss_fn(test_logits, y_test)
        # Update the learning rate based on the validation loss
        scheduler.step(test_loss)

        test_acc = accuracy(y_true=y_test,
                            y_pred=test_pred)
        
        if (epoch+1) % 100 == 0:
            print(f"Epoch: {epoch+1}, Loss: {loss:.5f}, Train Acc: {train_acc:.2f}%, Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")
```
